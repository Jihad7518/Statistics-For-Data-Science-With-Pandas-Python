# 1. Read two inputs from the user. Write a Python program to print their product. If the product is greater than 1000, then print their sum as well.

```python
number01 = int(input("Please Enter First Number: "))
number02 = int(input("Please Now Enter second Number for the operation: ")) # input() returns value as string, for that we need to use int or float first

product = number01 * number02 #product
sum = number01 + number02 #sum

if product > 1000: #condition
  print("Product of these two numbers are: ", product)
  print("Also, the sum of these two numbers are: ", sum)
else:
  print("Product of these two numbers are: ", product)
```

# 2. Write a Python program to find the area and perimeter of a circle. Read inputs from the user.

```python
radius = float(input("Enter the rdius of the circle: ")) # takes radius from user

area = 3.1416 * radius * radius # calculate area
perimeter = 2 * 3.1416 * radius # calculate perimeter

print("Area of the circle: ", area) #print area
print("Perimete of the circle: ", perimeter) #print the perimeter of the circle
```

# 3. Given a positive integer N (read from the user), write a Python program to calculate the value of the following series.
`1^2 + 2^2 + 3^2 + 4^2 + … + N^2`

```python
n = int(input("Enter an integer: ")) # takes integer from user
result = 0

for i in range(1, n+1, 1): # loop from 1 to n+1 increasing by 1
  result += i*i # sum of squares

print(result)
```

# 4. Given a positive integer N (read from the user), write a Python program to check if the number is prime or not.

```python
number = int(input("Enter a positive number please: ")) #takes integer from user
prime = False # define a varible for prime or not checking

if number <= 0:
  print("Please provide a positive integer!!!")
elif number == 1:
  print("Your given number is not a prime number!")
else:
  for i in range(2, (number//2)+1): #checking for factors
    if(number % i) == 0:
      prime = True # factors found so set prime to true
      break # as factors found, means it's not a prime, so no need for further checking

    if prime == True:
      print("Your given number is not a prime number!!!")
    else:
      print("Your given number is a prime number")
```

# 5. Given a list of numbers (hardcoded in the program), write a Python program to calculate the sum of the list. Do not use any built-in function.

```python
list1 = [10, 20, 30, 40, 50] #hardcoded list
sum = 0

for i in range(0, len(list1)): #iterate every element from the list and add them one by one
  sum = sum + list1[i]

print("Sum of all the list elements: ", sum)
```

# 8. Given a string, display only those characters which are present at an even index number. Read inputs from the user.

# 11. Given a two list of numbers (hardcoded in the program), create a new list such that new list should contain only odd numbers from the first list and even numbers from the second list. You can assume that the length of both list are the same.
